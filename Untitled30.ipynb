```python
from sklearn.preprocessing import OneHotEncoder
```

```python
df['new_pro']=df['procedural programming language'].cat.codes
df['new_fun']=df['functional programming language'].cat.codes
df['new_obj']=df['object oriented programming language'].cat.codes
df['new_scr']=df['scripting programming language'].cat.codes
df['new_log']=df['logic programming language'].cat.codes
df
```

```python
enc=OneHotEncoder()
```

```python
enc_df=pd.DataFrame(enc.fit_transform(df[['new_pro']]).toarray())
enc_df
```

```python
new_df=df.join(enc_df)
new_df
```
